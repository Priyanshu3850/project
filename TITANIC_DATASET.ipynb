```python
import pandas as pd
```

```python
titanic=pd.read_csv('/content/titanic_train.csv')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.info()
```

```python
port=pd.get_dummies(titanic.embarked,prefix='embarked')
port.head()
```

```python
titanic=titanic.join(port)
titanic.drop(['embarked'],axis=1,inplace= True)
titanic.head()
```

```python
titanic.sex=titanic.sex.map({'male':0,'female':1})
titanic.head()
```

```python
y=titanic.survived.copy()
x=titanic.drop(['survived'],axis=1)
```

```python
x.drop(['cabin','ticket','name','passenger_id'],axis=1,inplace=True)
```

```python
x.info
```

```python
x.isnull().sum()
```

```python
x.drop(['home.dest','body','boat',],axis=1,inplace=True)
```

```python
x
```

```python
x.isnull().sum()
```

```python
x.age.fillna(x.age.mean(),inplace=True)
x.head()
```

```python
x.isnull().sum()
```

```python
x.fare.fillna(x.fare.mean(),inplace=True)
x.head()
```

```python
x.isnull().sum()
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
